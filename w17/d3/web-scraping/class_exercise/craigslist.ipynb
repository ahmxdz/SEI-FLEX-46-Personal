```python
import requests
from bs4 import BeautifulSoup

# Target web page:
url = "https://en.wikipedia.org/wiki/List_of_English_football_champions"

# Establishing the connection to the web page:
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
result_tables = soup.find_all('table', {'class': 'wikitable'})
```

```python
print(len(result_tables))
```

```python
result_tables[2]
```

```python
for result_row in result_tables[2].tbody.find_all('tr'):
    result = result_row.find_all('td')
    if len(result):
        print(f'{result[0].text} -- {result[1].text} -- {result[4].text}')
```

```python
#use css selector
for result_row in result_tables[2].find('tbody').find_all('tr'):
    print(result_row.select_one(':nth-child(1)').text)
    print(result_row.select_one(':nth-child(2)').text)
```
